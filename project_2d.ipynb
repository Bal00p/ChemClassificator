```python
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader, TensorDataset
from torchvision.transforms import ToTensor
```

```python
from sklearn.preprocessing import MinMaxScaler

X = data.drop(columns=['MolWt'])
y = data['MolWt']

# Преобразование данных в диапазон от 0 до 1
# scaler = MinMaxScaler()
# X_scaled = scaler.fit_transform(X)
# y = scaler.fit_transform(y.values.reshape(-1, 1))

X_tensor = torch.tensor(X.values, dtype=torch.float32)
y_tensor = torch.tensor(y.values, dtype=torch.float32)
```

```python
import torch.nn.functional as F

# x_train, x_test, y_train, y_test = train_test_split(X_tensor, y_tensor, random_state=42, train_size=0.7, shuffle=True)

# Определение размера батча
batch_size = 4  # Измените это значение на свое усмотрение

# Преобразование данных в тензоры PyTorch
x_train_tensor = torch.tensor(X_train, dtype=torch.float32)
y_train_tensor = torch.tensor(y_train, dtype=torch.float32).unsqueeze(1)

# Создание DataLoader для обучения с новым размером батча
train_dataset = TensorDataset(x_train_tensor, y_train_tensor)
train_loader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True)

n_samples, n_features = X_train.shape[0], X_train.shape[1]
input_size = n_features
output_size = 1

# class LinearRegression(nn.Module):
#     def __init__(self, input_size, output_size, dropout_prob=0.5):
#         super(LinearRegression, self).__init__()
#         self.linear = nn.Linear(input_size, output_size)
#         self.dropout = nn.Dropout(dropout_prob)
#         self.sigmoid = nn.Sigmoid()  # Добавлено для ограничения значений в диапазоне от 0 до 1

#     def forward(self, x):
#         x = self.dropout(F.relu(self.linear(x)))
#         x = self.sigmoid(x)  # Применение сигмоиды для ограничения значений в диапазоне от 0 до 1
#         return x

class NeuralNetwork(nn.Module):
    def __init__(self, input_size, hidden_size1, hidden_size2, output_size, dropout_prob=0.9):
        super(NeuralNetwork, self).__init__()
        self.fc1 = nn.Linear(input_size, hidden_size1)
        self.fc2 = nn.Linear(hidden_size1, hidden_size2)
        self.fc3 = nn.Linear(hidden_size2, output_size)
        self.dropout = nn.Dropout(dropout_prob)
        self.relu = nn.ReLU()

    def forward(self, x):
        x = F.relu(self.fc1(x))
        x = self.dropout(x)
        x = F.relu(self.fc2(x))
        x = self.dropout(x)
        x = self.fc3(x)
        x = self.relu(x)
        return x
```

```python
y_train_tensor
```

```python
model = NeuralNetwork(input_size, 128, 64, output_size)

# Выбор функции потерь MSE
criterion = nn.MSELoss()

# Определение оптимизатора
optimizer = optim.Adam(model.parameters(), lr=0.01)#, momentum=0.9)

# Обучение модели
epochs = 20

for epoch in range(epochs):
    running_loss = 0.0

    for inputs, labels in train_loader:
        # Обнуление градиентов
        optimizer.zero_grad()

        # Forward pass
        outputs = model(inputs)
        loss = criterion(outputs, labels)

        # Backward pass и оптимизация
        loss.backward()
        optimizer.step()

        running_loss += loss.item() * inputs.size(0)

    # Вычисление среднего значения потерь на эпоху
    epoch_loss = running_loss / len(train_loader.dataset)

    print(f'Epoch [{epoch + 1}/{epochs}], Loss: {epoch_loss:.4f}')
```

```python
y_torch = torch.tensor(y, dtype=torch.float32).unsqueeze(1)
x_torch = torch.tensor(X.values, dtype=torch.float32)

class LinearRegression(nn.Module):
    def __init__(self):
        super().__init__()
        self.layer = nn.Linear(167,1)
    def forward(self, x_torch):
        return self.layer(x_torch)

model = LinearRegression()

loss_fn = nn.MSELoss()
learning_rate = 0.01
optimizer = torch.optim.Adam(params= model.parameters(), lr = learning_rate)

loss_list=[]
epochs = 2000
for epoch in range(epochs):
    model.train()

    y_pred = model(x_torch)

    loss = loss_fn(y_pred, y_torch)

    optimizer.zero_grad()

    loss.backward()

    optimizer.step()

    loss_list.append(loss.data)

    if(epochs % 50 == 0):
        print('epoch {}, loss {}'.format(epoch, loss.data))

    model.eval()
```
